```python
# split X and Y into training and testing sets

from sklearn import preprocessing
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state =0)

print(X_train.shape, X_test.shape)
```

```python
# using the standard scaler for feature scaling

scaler = preprocessing.StandardScaler()

# fit transform is used in the training data so that we can scale the training data and 
# learn the scaling parameters of the data

X_train_standard = scaler.fit_transform(X_train)

X_test_standard = scaler.transform(X_test)
```

```python
# LINEAR REGRESSION
#import the class
from sklearn.linear_model import LinearRegression

#initiate the model (using default parameters)
ligreg = LinearRegression()

#fit the model with training data
ligreg.fit(X_train,Y_train)

# predict the output for test set 
Y_pred = ligreg.predict(X_test)
Y_pred
new_prediction = ligreg.predict(scaler.transform(np.array([[100,0,13,50,50]])))
print("Prediction score:" , new_prediction)

#compute R2 score - coefficient of determination
# Best possible score is 1.0
score = ligreg.score(X_test,Y_test)
print("R-squared value:" , score)
```
